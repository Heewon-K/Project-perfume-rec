```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
```

```python
# Load Pretrained Word2Vec
embed = hub.load("https://tfhub.dev/google/Wiki-words-250-with-normalization/2")
```

```python
import pandas as pd
cat = pd.read_csv('../data/concat_category.csv')

name = list(cat['name'])
for i in range(len(name)):
    name[i] = name[i].replace("_", " ")
name
```

```python
name2 = ['citrus', 'fruits', 'flowers', 'white flowers', 'herbs', 'spices', 'sweets', 'woods', 'resins', 'musk', 'beverages', 'synthetic']
```

```python
embeddings = embed(name2)
```

```python
print(embeddings[0]) 
```

```python
tf.convert_to_tensor(embeddings, dtype=None, dtype_hint=None, name=None)
```

```python
layer = tf.keras.layers.Normalization(axis=None)
layer.adapt(embeddings)
print(embeddings.shape)
```

```python
# Compute similarity matrix. Higher score indicates greater similarity.
sim = []
for i in range(len(name2)):
    for j in range(i,len(name2)):
        sim.append(["(" + name2[i] + "," + name2[j] + ")", np.inner(embeddings[i], embeddings[j])])
sim
```

```python
# Python program to store list to file using pickle module
import pickle

with open('../data/sim.pkl', "wb") as f:
    pickle.dump(sim, f)
with open('../data/sim.pkl', "rb") as f:
    sim_l = pickle.load(f)
```

```python
mtx = pd.DataFrame(columns = name2, index = name2)
mtx.info
```

```python
# citrus 채우기
for i in range(0,12):
    mtx['citrus'][i] = sim[i][1]
# fruits 채우기
for i in range(0,11):
    mtx['fruits'][i+1] = sim[i+12][1]
# flowers 채우기
for i in range(0,10):
    mtx['flowers'][i+2] = sim[i+23][1]
# white flowers 채우기
for i in range(0,9):
    mtx['white flowers'][i+3] = sim[i+33][1]
# herbs
for i in range(0,8):
    mtx['herbs'][i+4] = sim[i+42][1]
# spices
for i in range(0,7):
    mtx['spices'][i+5] = sim[i+50][1]
# sweets
for i in range(0,6):
    mtx['sweets'][i+6] = sim[i+57][1]
# woods
for i in range(0,5):
    mtx['woods'][i+7] = sim[i+63][1]
# resins
for i in range(0,4):
    mtx['resins'][i+8] = sim[i+68][1]
# musk
for i in range(0,3):
    mtx['musk'][i+9] = sim[i+72][1]
# beverages
for i in range(0,2):
    mtx['beverages'][i+10] = sim[i+75][1]
# synthetic
for i in range(0,1):
    mtx['synthetic'][i+11] = sim[i+77][1]
```

```python
mtx = mtx.fillna(0)
mtx_t = mtx.T.fillna(0)
mtx_m = mtx + mtx_t
for i in range(0,12):
    mtx_m.iloc[i,i] = (mtx_m.iloc[i,i]/2)
mtx_m
```

```python
from sklearn import preprocessing

x = mtx_t.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
scaled_m = pd.DataFrame(x_scaled)
scaled_m
```

```python
scaled_c = scaled_m + scaled_m.T

for i in range(0,12):
    scaled_c.iloc[i,i] = (scaled_c.iloc[i,i]/2)
scaled_c
```

```python
scaled_c.index, scaled_c.columns = [name2, name2]
scaled_c
```

```python
scaled_c.to_csv('../data/similarity_matrix.csv')
df = pd.read_csv('../data/similarity_matrix.csv', index_col = 0)
df
```
